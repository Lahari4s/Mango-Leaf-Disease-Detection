```python
from tensorflow.keras.models import load_model
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input as efficientnet_preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
```

```python
# Preprocess the image
def preprocess_image(image_path, target_size=(224, 224)):
    """
    Preprocess the input image for EfficientNetB0:
    - Load image
    - Resize to target size
    - Convert to array
    - Use EfficientNetB0 preprocess_input for normalization
    """
    try:
        # Load image with target size
        image = load_img(image_path, target_size=target_size)

        # Convert image to numpy array
        image_array = img_to_array(image)

        # Preprocess using EfficientNetB0's preprocess_input
        image_array = efficientnet_preprocess_input(image_array)

        # Add batch dimension
        image_array = np.expand_dims(image_array, axis=0)

        return image_array
    except Exception as e:
        print(f"Error in preprocessing: {e}")
        return None
```

```python
# Path to your EfficientNetB0 .h5 model
model_path = "mango_leaf_disease_efficientnetb0.h5"
# Path to your test image
image_path = r"D:\projects\external\mango_disease_clasification\input\mango-leaf-disease-dataset\Gall Midge\IMG_20211106_161105 (Custom).jpg"

# Load the model
try:
    model = load_model(model_path)
    print("Model loaded successfully!")
except Exception as e:
    print(f"Error loading model: {e}")

# Preprocess the image
input_image = preprocess_image(image_path)

if input_image is not None:
    # Predict with the model
    predictions = model.predict(input_image)

    # Assuming your model has class names defined
    class_names = ['Anthracnose', 'Bacterial Canker', 'Cutting Weevil', 'Die Back', 'Gall Midge', 'Healthy', 'Powdery Mildew', 'Sooty Mould']

    # Get the class with the highest probability
    predicted_class = class_names[np.argmax(predictions)]
    print(f"Predicted class: {predicted_class}, Confidence : {np.max(predictions)*100} %")
else:
    print("Failed to preprocess image.")
```
